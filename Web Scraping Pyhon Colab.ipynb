Step 2. Scrape the comments, comment ID, comment score, and date created

```python
comment_body_list = [] #creating an empty list to store the body of each comment

for comment in lgbtq_post.comments:	#starting a for-loop to iterate through the comment forest of the Reddit post 
	if type(comment) == MoreComments:	#logical operator checks if the comment is a "MoreComments" object (contains a link to load more comments). If yes, the loop skips it and continues through.  
		continue

	comment_body_list.append(comment.body) #adding the body of each comment to comment_body_list using append() function

#creating a Pandas dataframe
comments_body_df = pd.DataFrame(comment_body_list, columns=['comment_body']) #passing comment_body_list and its column name 'comment_body' using pd.DataFrame() method to convert to a Pandas dataframe 
comments_body_df #using head() function to display the first 5 rows of the new dataframe 
```

Scraping nested comments

```python
comment_data_list = []  #create a new empty list to store the comment data

lgbtq_post.comments.replace_more(limit=None) #makes sure that all the comments are loaded before the for-loop iterates through each comment
for comment in lgbtq_post.comments.list(): #starting a for loop that iterates through each comment in the lgbtq_post Reddit post object 
 comment_data_list.append([comment.body, comment.score, comment.id, comment.created]) #appends a list of properties to the comment_data_list list using append()

#converting the 'post_comments' list to dataframe and naming the columns.
comments_data_df = pd.DataFrame(comment_data_list, columns=['comment_body', 'comment_score', 'comment_ID', 'comment_created']) #converting comments_data_list and its columns into a Pandas dataframe
comments_data_df
```

Converting the date

```python
from datetime import datetime #import datetime module from Python standard library

comments_data_df['date_time'] = pd.to_datetime(comments_data_df['comment_created'],  unit='s') #creating a new column 'date_time' by converting the timestamp values in comment_created column to a Pandas datetime format. unit parameter tells Pandas that the input values are in seconds.  
comments_data_df['date'] = comments_data_df['date_time'].dt.strftime('%Y-%m-%d') #creating another new column 'date' by extracting the date portion from 'date_time' via strftime() method. The string %Y-%m-%d tells Pandas that the output format should be "YYYY-MM-DD"

comments_data_df.head() #view the first 5 rows of the updated Dataframe
```

Step 3. Delete duplicated rows and save the file in CSV format

```python
comments_data_df.drop_duplicates(inplace=True) #removing duplicate rows in comments_data_df; inplace=True parameter makes sure that the rows are dropped from the original dataframe instead of creating a new one
comments_data_df.shape #print the shape of the dataframe (rows x columns)

#save as CSV file
comments_data_df.to_csv('comments.csv', index=False) #saving the created dataframe to a CSV, preventing the index column from being included in the file with index=False parameter
comments_data_df.to_csv
```
